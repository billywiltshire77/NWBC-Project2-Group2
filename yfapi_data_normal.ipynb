```python
# Code to normaize the rataing of the stock from the Yahoo Finance API
import pandas as pd
import os  
```

```python
# Store filepath in a variable
file_one = "yfapi_data.csv"
```

```python
# Read our Data file with the pandas library
file_one_df = pd.read_csv(file_one, encoding="UTF-8")
file_one_df.head(5)
```

```python
file_one_df[['num_rating','rating']] = file_one_df['avg_analyst_rating'].str.split('-', expand=True)
file_one_df = file_one_df.drop('avg_analyst_rating', axis=1)
```

```python
file_one_df.head()
```

```python
file_one_df['norm_rating'] = ""
file_one_df.head()
```

```python
file_one_df['rating'] = file_one_df['rating'].str.strip()
file_one_df.head()
```

```python
for index, row in file_one_df.iterrows():
    rating = row['rating']

    if rating == 'Strong Buy':
        norm_rating = 5
    elif rating == 'Buy':
        norm_rating = 4
    elif rating == 'Hold':
        norm_rating = 3
    elif rating == 'Under Perform':
        norm_rating = 2
    else:
        norm_rating = 1 

    file_one_df.loc[index, 'norm_rating'] = norm_rating

file_one_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
output_path = os.path.join("yfapi_data_nrmal.csv")
file_one_df.to_csv(output_path, index=False, header=True)
print(output_path)
```
